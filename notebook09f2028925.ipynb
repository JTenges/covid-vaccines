```python
data = pd.read_csv('country_vaccinations.csv')
data = data.groupby(['country', 'iso_code'])
data = data.max() # Use latest data
data = data.reset_index()
data.sort_values('country')
data.head(10)
```

```python
ax = data.plot.bar(x='country', y='total_vaccinations', rot=90, figsize=(30, 10))

ax.set_xlabel("Country")
ax.set_ylabel("Total Vaccinations")
ax.set_title("Total Vaccinations for each country")
ax.get_legend().remove()
plt.show()
```

```python
ax = data.plot.bar(x='country', y='total_vaccinations_per_hundred', rot=90, figsize=(30, 10))

ax.set_xlabel("Country")
ax.set_ylabel("Total Vaccinations Per Hundred")
ax.set_title("Total Vaccinations Per Hundred for each Country")
ax.get_legend().remove()
plt.show()
```

```python
fig = px.choropleth(data, locations='iso_code',
                    color='total_vaccinations',
                    hover_name='country',
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

```python
fig = px.choropleth(data, locations='iso_code',
                    color='total_vaccinations_per_hundred',
                    hover_name='country',
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```
